```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-20, 20)

y = 2 * x + 3

plt.figure(figsize=(8, 6), facecolor= 'lightblue', edgecolor ='b')
plt.plot(x, y,  color='red')

plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('Graph of y = 2x + 3', fontsize=14)


plt.axhline(0, color='green')
plt.axvline(0, color='green')
plt.legend(['y = 2x + 3'])

plt.show()
```

```python
# 2. Display the population of 5 different countries
countries = ['China', 'India', 'United States', 'Indonesia', 'Pakistan']
populations = [1441, 1393, 331, 273, 225]

plt.figure(figsize=(10, 6))
plt.bar(countries, populations, color='skyblue')

plt.title('Population of Different Countries ', fontsize=16)
plt.xlabel('Countries', fontsize=14)
plt.ylabel('Population ', fontsize=14)

plt.grid(axis='y')
plt.show()
```

```python
y = x*x
plt.figure(figsize = (12,8),facecolor='orange')
plt.plot(x,y,color='green',linestyle='-.', linewidth=1.8)

plt.title('Graph of y=x^2',fontsize=14)
plt.xlabel('x')
plt.ylabel('y')

plt.grid(True)     #to add grid

plt.legend(['y=x^2'])
plt.show()
```

```python
sin_x = np.sin(x)
cos_x = np.cos(x)
tan_x = np.tan(x)
cot_x = 1/tan_x

fig, axs = plt.subplots(2,2, figsize = (12,10))

axs[0,0].plot(x,sin_x,color='violet')
axs[0,0].set_title('Sine Function')
axs[0,0].set_xlabel('x')
axs[0,0].set_ylabel('sin(x)')
axs[0,0].grid(True)
axs[0,0].axhline(0, linewidth=0.8)
axs[0,0].axvline(0,linewidth=0.8)
axs[0,0].legend(['sin(x)'])

axs[0,1].plot(x,cos_x,color='blue')
axs[0,1].set_title('Cosine Function')
axs[0,1].set_xlabel('x')
axs[0,1].set_ylabel('cos(x)')
axs[0,1].grid(True)
axs[0,1].axhline(0, linewidth=0.8)
axs[0,1].axvline(0,linewidth=0.8)
axs[0,1].legend(['cos(x)'])

axs[1,0].plot(x,tan_x,color='green')
axs[1,0].set_title('Tangent Function')
axs[1,0].set_xlabel('x')
axs[1,0].set_ylabel('tan(x)')
axs[1,0].grid(True)
axs[1,0].axhline(0, linewidth=0.8)
axs[1,0].axvline(0,linewidth=0.8)
axs[1,0].legend(['tan(x)'])

axs[1,1].plot(x,cot_x,color='red')
axs[1,1].set_title('Cotangent Function')
axs[1,1].set_xlabel('x')
axs[1,1].set_ylabel('cot(x)')
axs[1,1].grid(True)
axs[1,1].axhline(0, linewidth=0.8)
axs[1,1].axvline(0,linewidth=0.8)
axs[1,1].legend(['cot(x)'])

plt.show()
```

```python
companies = ['Google', 'Microsoft', 'Tata', 'IBM', 'Accenture']
revenues =  [14796.08, 15463.13, 10557.35, 7250.08, 3442.21]

plt.figure(figsize = (10,8),facecolor='lightblue',edgecolor='b')
plt.pie(revenues, labels=companies, autopct='%0.1f%%')

plt.title('Revenue Generation by Companies')

plt.axis('equal')        #ensure pie chart is circular

plt.show()
```

```python
frequency = 1
amplitude = 1
t = np.linspace(0, 10, 1000)
cos_wave = amplitude * np.cos(2 * np.pi * frequency * t)

plt.figure(figsize=(12,8))
plt.plot(t,cos_wave, color = 'blue')

plt.title('Cosine Wave Over Time')
plt.xlabel('Time')
plt.ylabel('Amplitude')

plt.grid(True)
plt.legend(['Cosine Wave'])

plt.axhline(0,color='black')
plt.axvline(0,color='black')
plt.show()
```
