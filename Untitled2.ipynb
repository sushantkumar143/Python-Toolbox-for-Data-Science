```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

1. Display Missing Values before handling
2. Replace missing "Age" values with median Age
3. Drop the cabin column
4. Display missing values after handelling.
5. Display first few rows after the cleaned datasets.

```python
df = pd.read_csv('/content/Titanic-Dataset.csv')
df
```

```python
# FILLING MISSING VALUES WITH MEN (FOR NUMERICAL) AND MODE (FOR CATEGORIAL)
imputer = SimpleImputer(strategy='median')
df['Age'] = imputer.fit_transform(df[['Age']])
df['Age'] = df['Age'].astype(int)
df.head(10)
```

```python
df.drop(columns=['Cabin'], inplace=True)
df.head(10)
```
