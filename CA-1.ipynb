```python
# QUESTION - 6

df['Gender'] = df['Gender'].str.lower()
df['Country'] = df['Country'].str.capitalize()
df['Nodule_Size'] = df['Nodule_Size'].astype(int)
df.head(5)
```

```python
var1 = df.groupby(['Country'])['Nodule_Size'].mean()
plt.figure(figsize=(10,5), facecolor="aliceblue")
var1.plot(kind='line', marker="o")
plt.xlabel('Country')
plt.legend("Country")
plt.ylabel('Nodule_Size')
plt.title('Country vs Nodule_Size')
plt.show()
```

```python
# QUESTION - 7
#
var2 = df.groupby(['Country'])['TSH_Level'].sum()

plt.figure(figsize=(10,5), facecolor="aliceblue")
plt.bar(var2.index, var2.values)
plt.xticks(rotation=50)
plt.xlabel('Country')
plt.ylabel('TSH_Level')
plt.title('Country vs TSH_Level')
# plt.show()
```
